```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print("First 5 rows of the DataFrame:")
print(df.head())
skewness_values=df.skew()
print("\nSkewness values:")
print(skewness_values)
kurtosis_values=df.kurtosis()
print("\nKurtosis values:")
print(kurtosis_values)
```

```python
import pandas as pd
import numpy as np

# Load the tips dataset
df_tips = pd.read_csv("tips.csv")

print("First 5 rows of the tips DataFrame:")
print(df_tips.head())

# Select only numerical columns for skewness and kurtosis calculations
numerical_df_tips = df_tips.select_dtypes(include=[np.number])

skewness_values_tips = numerical_df_tips.skew()
print("\nSkewness values for tips dataset:")
print(skewness_values_tips)

kurtosis_values_tips = numerical_df_tips.kurtosis()
print("\nKurtosis values for tips dataset:")
print(kurtosis_values_tips)
```
